# Demand Forecasting & Scenario Modeling

This notebook builds weekly demand forecasts for a travel platform using historical booking data. The objective is to generate interpretable, seasonality-aware forecasts that can be translated into revenue, capacity utilization, and financial planning insights.

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX

from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv(
    "cleaned_hotel_bookings.csv",
    parse_dates=["arrival_date", "week_start"]
)

df.head()
```

```python
weekly = (
    df.groupby(["week_start", "hotel"])
      .agg(
          net_bookings=("net_booking", "sum"),
          room_nights=("total_stay_nights", "sum"),
          avg_adr=("adr", "mean")
      )
      .reset_index()
)

weekly.head()
```

```python
weekly_pivot = (
    weekly.pivot(index="week_start", columns="hotel", values="room_nights")
          .sort_index()
)

weekly_pivot.head()
```

```python
# Seasonality Validation 
# City Hotel Decomposition
city_series = weekly_pivot["City Hotel"].dropna()

decomp_city = seasonal_decompose(city_series, model="additive", period=52)

decomp_city.plot()
plt.show()
```

```python
# Resort Hotel Decomposition
resort_series = weekly_pivot["Resort Hotel"].dropna()

decomp_resort = seasonal_decompose(resort_series, model="additive", period=52)

decomp_resort.plot()
plt.show()
```

Seasonal decomposition confirms strong annual seasonality (52-week cycle) in both City and Resort demand,
with higher amplitude observed in Resort hotels. This validates the use of seasonal forecasting models.

```python
# Trainâ€“Test Split (Time-Series Correct)
train_end = int(len(city_series) * 0.85)

city_train, city_test = city_series[:train_end], city_series[train_end:]
resort_train, resort_test = resort_series[:train_end], resort_series[train_end:]
```

```python
len(city_series), len(resort_series)
```

```python
weekly_pivot = weekly_pivot.sort_index()
weekly_pivot = weekly_pivot.asfreq("W-MON")
```

```python
weekly_pivot.isna().sum()
```

```python
city_series = weekly_pivot["City Hotel"]
resort_series = weekly_pivot["Resort Hotel"]
```

```python
# Model 1: Exponential Smoothing (ETS)
# City Hotel

ets_city = ExponentialSmoothing(
    city_series,
    trend="add",
    damped_trend=True,
    seasonal=None
).fit(optimized=True)

city_ets_forecast = ets_city.forecast(len(city_test))
```

```python
# Resort Hotel
ets_resort = ExponentialSmoothing(
    resort_series,
    trend="add",
    seasonal="add",
    seasonal_periods=52
).fit(optimized=True)

resort_ets_forecast = ets_resort.forecast(len(resort_test))
```

```python
# Model 2: SARIMA (Seasonal ARIMA)
# City Hotel
sarima_city = SARIMAX(
    city_series,
    order=(1,1,0),
    seasonal_order=(1,0,0,52),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

city_sarima_forecast = sarima_city.forecast(len(city_test))
```

```python
# Resort Hotel
sarima_resort = SARIMAX(
    resort_series,
    order=(1,1,0),
    seasonal_order=(1,0,0,52),
    enforce_stationarity=False,
    enforce_invertibility=False
).fit(disp=False)

resort_sarima_forecast = sarima_resort.forecast(len(resort_test))
```

```python
# Rolling Validation (Finance-credible accuracy)
def mape(actual, forecast):
    return mean_absolute_percentage_error(actual, forecast) * 100


results = pd.DataFrame({
    "Model": ["ETS City", "SARIMA City", "ETS Resort", "SARIMA Resort"],
    "MAPE (%)": [
        mape(city_test, city_ets_forecast),
        mape(city_test, city_sarima_forecast),
        mape(resort_test, resort_ets_forecast),
        mape(resort_test, resort_sarima_forecast),
    ]
})

results
```

```python
# Final forecast
forecast_horizon = 52

final_city_model = ets_city if results.loc[0,"MAPE (%)"] < results.loc[1,"MAPE (%)"] else sarima_city
final_resort_model = ets_resort if results.loc[2,"MAPE (%)"] < results.loc[3,"MAPE (%)"] else sarima_resort

city_forecast = final_city_model.forecast(forecast_horizon)
resort_forecast = final_resort_model.forecast(forecast_horizon)
```

```python
# Scenario Generation (Finance Layer Input)
scenario_df = pd.DataFrame({
    "Week": city_forecast.index,
    "City_Base": city_forecast,
    "City_Peak": city_forecast * 1.15,
    "City_Shock": city_forecast * 0.90,
    "Resort_Base": resort_forecast,
    "Resort_Peak": resort_forecast * 1.15,
    "Resort_Shock": resort_forecast * 0.90
})

scenario_df.head()
```

The initial city Hotel forecast shows a pronounced downward level shift immediately after the forecast origin, with projected demand stabilizing significantly below recent historical levels. This behavior indicates that the model is over-weighting long-term historical averages and seasonal structure relative to the most recent demand regime.

City demand exhibits multiple structural changes and sharp disruptions, suggesting that it is driven more by trend and external shocks than by stable seasonal patterns. Applying a fully seasonal ETS model in this context causes the forecast to revert toward a long-run mean rather than anchoring to the latest observed demand. This highlights the risk of imposing strong seasonality assumptions on series with heterogeneous demand drivers.

As a result, the initial model understates near-term City demand and would lead to overly conservative revenue and capacity planning decisions if used without adjustment.

## Seasonality Validation and Decomposition Approach

Classical seasonal decomposition was initially applied to validate the presence of seasonality in weekly hotel demand. This method decomposes the time series into trend, seasonal, and residual components under the assumption of fixed, repeating seasonal patterns and a globally smooth trend. While suitable for exploratory analysis, this approach is sensitive to demand shocks and assumes stable seasonality across years.

Subsequently, STL (Seasonal and Trend decomposition using LOESS) was employed to address the limitations of classical decomposition. STL allows both trend and seasonality to evolve locally over time and, when used with robust estimation, reduces the influence of outliers such as cancellation-driven spikes. This makes STL more appropriate for noisy, real-world demand series, particularly for City Hotels where seasonality is present but unstable.

Classical decomposition was therefore used for diagnostic validation, while STL served as the foundation for forecasting by enabling flexible separation of trend, seasonal structure, and residual uncertainty.

# Rolling Validation
Rolling (walk-forward) validation can be used to evaluate the STL + ETS model under realistic forecasting conditions. The model demonstrates stable short-term accuracy while preserving seasonal structure, supporting its use as a demand forecasting approach for City Hotels.

```python
# Rolling STL + ETS Function
def rolling_stl_ets_forecast(
    series,
    initial_train_size=78,   # ~1.5 years
    seasonal_period=52
):
    actuals = []
    forecasts = []

    for t in range(initial_train_size, len(series)):
        train = series[:t]

        # STL decomposition on training data
        stl = STL(train, period=seasonal_period, robust=True)
        stl_result = stl.fit()

        trend = stl_result.trend.dropna()
        seasonal = stl_result.seasonal

        # ETS on trend component
        trend_model = ExponentialSmoothing(
            trend,
            trend="add",
            seasonal=None
        ).fit(optimized=True)

        # 1-step ahead trend forecast
        trend_fc = trend_model.forecast(1)

        # Reuse most recent seasonal value
        seasonal_fc = seasonal.iloc[-seasonal_period]

        forecast = trend_fc.values[0] + seasonal_fc

        actuals.append(series.iloc[t])
        forecasts.append(forecast)

    return np.array(actuals), np.array(forecasts)
```

```python
# Rolling Validation for STL+ETS model
city_actuals_stl, city_forecasts_stl = rolling_stl_ets_forecast(
    city_series
)

city_mape_stl = mean_absolute_percentage_error(
    city_actuals_stl,
    city_forecasts_stl
) * 100

print(f"City Hotel STL + ETS Rolling MAPE: {city_mape_stl:.2f}%")
```

Rolling (walk-forward) validation was used to evaluate the STL + ETS model under realistic
forecasting conditions. The model demonstrates stable short-term accuracy while preserving
seasonal structure, supporting its use as a demand forecasting approach for City Hotels.

Rolling one-step-ahead validation shows that the STL + ETS model tracks the direction of City Hotel demand reasonably well, with larger errors during volatile periods. The observed MAPE of 22.5% is acceptable given demand uncertainty and cancellation-driven variability, supporting the modelâ€™s use for short-term planning and scenario analysis.

```python
# Rolling validation for Flat Model
def rolling_flat_forecast(series, initial_train_size=78):
    actuals = []
    forecasts = []

    for t in range(initial_train_size, len(series)):
        train = series[:t]

        model = ExponentialSmoothing(
            train,
            trend=None,
            seasonal=None
        ).fit(optimized=True)

        forecast = model.forecast(1)[0]

        actuals.append(series.iloc[t])
        forecasts.append(forecast)

    return np.array(actuals), np.array(forecasts)
```

```python
city_actuals_flat, city_forecasts_flat = rolling_flat_forecast(city_series)

city_mape_flat = mean_absolute_percentage_error(
    city_actuals_flat,
    city_forecasts_flat
) * 100

print(f"City Hotel Flat Model Rolling MAPE: {city_mape_flat:.2f}%")
```

Rolling validation shows that a flat local-level model achieves lower one-step-ahead error,
reflecting the stability advantage of level-based forecasts in short horizons. However,
for medium-term planning and scenario analysis, the STL + ETS model is preferred as it
captures observed seasonal structure without imposing long-term trends, providing more
realistic demand trajectories.
